```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# 1. Cargar la imagen
image_path = "/Users/juanreyes/Documents/Juan Sebastian/MAIA/Prueba_clas_img/image.jpeg"  # Reemplzalo con la ruta real
image = cv2.imread(image_path)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)  # Convertir de BGR a RGB (OpenCV usa BGR por defecto)

# 2. Redimensionar la imagen (opcional)
image = cv2.resize(image, (200, 200))  # Puedes cambiar el tamao segn lo necesites

# 3. Convertir la imagen a un vector de pxeles
pixels = image.reshape((-1, 3))  # Convertir a una matriz de pxeles (cada fila es un pxel con 3 valores RGB)

# 4. Aplicar K-Means
k = 5  # Nmero de clusters
kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
kmeans.fit(pixels)

# 5. Asignar cada pxel a su cluster
segmented_pixels = kmeans.cluster_centers_[kmeans.labels_]
segmented_image = segmented_pixels.reshape(image.shape).astype(np.uint8)

# 6. Visualizar resultados
fig, ax = plt.subplots(1, 2, figsize=(10, 5))
ax[0].imshow(image)
ax[0].set_title("Imagen Original")
ax[0].axis("off")

ax[1].imshow(segmented_image)
ax[1].set_title(f"Imagen Segmentada con K={k}")
ax[1].axis("off")

plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# 1 Cargar la imagen
image_path = "/Users/juanreyes/Documents/Juan Sebastian/MAIA/Prueba_clas_img/2.jpeg"  # Reemplzalo con la ruta real
image = cv2.imread(image_path)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)  # Convertir de BGR a RGB

# 2 Redimensionar (opcional) para mejorar rendimiento
image = cv2.resize(image, (200, 200))  

# 3 Convertir la imagen a un vector de pxeles
pixels = image.reshape((-1, 3))  # Cada fila es un pxel con valores RGB

# 4 Aplicar K-Means para encontrar los colores dominantes
k = 3  # Nmero de clusters (colores)
kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
kmeans.fit(pixels)

# 5 Obtener los centroides (colores representativos)
centroids = np.round(kmeans.cluster_centers_).astype(int)  # Convertimos a valores enteros RGB
print("Colores representativos (RGB):", centroids)

# 6 Mostrar los colores representativos
def plot_colors(centroids):
    """Funcin para mostrar los colores obtenidos por K-Means."""
    plt.figure(figsize=(8, 2))
    plt.axis("off")
    plt.title("Colores Dominantes")

    # Crear una imagen de los colores en un rectngulo
    color_rect = np.zeros((50, 300, 3), dtype=np.uint8)
    step = 300 // len(centroids)  # Dividir el espacio entre los colores

    for i, color in enumerate(centroids):
        color_rect[:, i * step:(i + 1) * step] = color

    plt.imshow(color_rect)
    plt.show()

# Mostrar los colores dominantes
plot_colors(centroids)
```
